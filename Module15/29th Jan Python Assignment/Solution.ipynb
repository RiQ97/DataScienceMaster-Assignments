Q2. Which type of programming does Python support?


A2. Python is a multi-paradigm programming language, which means that it supports multiple programming paradigms. Few of them are as follow:

1. Procedural programming: A programming paradigm where code is written in a sequence. This is based on the use of functions, variables, and control flow statements such as if-else, for, and while.

2. Object-oriented programming (OOP): A programming paradigm that focuses on the use of objects to represent real-world entities and their interactions. This is based on the use of classes, objects, inheritance, encapsulation, and polymorphism.

3. Functional programming: A programming paradigm that emphasizes the use of functions as first-class citizens. This is based on the use of higher-order functions, lambda functions, and the avoidance of mutable state.

4. Aspect-oriented programming (AOP): A programming paradigm that focuses on the separation of concerns in a program. This is based on the use of decorators, which allow us to add new functionality to existing functions and classes.

Q6. Name a few blocks of code used to define in Python language?


A6. A program is constructed from various code blocks, each of which serves a specific purpose. Here are some of the key code blocks used in Python:

1. Module: A module is a file containing Python code that can be imported and used in other programs. It typically defines functions, classes, and variables. They are executed only once when imported.

2. Function Body: A function body contains the actual implementation of a function. It consists of a sequence of statements that execute when the function is called. It can be executed many times.

3. Class Definition: A class definition defines a blueprint for creating objects (instances). It includes attributes and methods that operate on those attributes. It define the behavior of objects.

4. Interactive Commands: When you type commands interactively in the Python interpreter (REPL), each command is treated as a code block. These commands are executed immediately.

5. Script File: A script file is a Python file executed from the command line or as standard input to the interpreter. It is also considered a code block.

6. Script Command: A command specified on the interpreter command line using the -c option is also a code block.

7. String Argument for eval() and exec(): The string argument passed to the built-in functions eval() and exec() is treated as a code block.

Q13. How is memory managed in Python?

A13. Memory management in Python is handled automatically by the Python interpreter, which includes several components to manage memory efficiently:

1. Garbage Collection: Python uses a garbage collector to de-allocate memory that is no longer in use. This process helps to free up memory for other objects and prevents memory leaks.
2. Reference Counting: Python internally keeps track of the number of references to each object in memory. When an object’s reference count drops to zero, meaning no references to the object exist, it is eligible for garbage collection.
3. Memory Pools: Python uses a system of memory pools for managing the allocation of memory to objects. This system helps to minimize the overhead of memory allocation and de-allocation.
4. Blocks and Arenas: Memory for Python objects is allocated in blocks, which are grouped into larger units called arenas. This structure helps to manage memory in a way that reduces fragmentation.

The combination of these mechanisms ensures that Python programs can run efficiently without the programmer needing to manually manage memory.

Q14. How to install Python on windows and set path variables?

A14. To install Python on Windows and set the PATH variables, you can follow these general steps:

During the installation, ensure to check the option that says “Add Python to PATH” at the bottom of the installation window. This will automatically set the PATH variable for you.

Manual PATH Setup:
Find the Python installation directory (usually C:\Users\<USER>\AppData\Local\Programs\Python\Python<version>).
Copy the path to the directory where python.exe is located.
Open the Start Search, type in “Environment Variables,” and select “Edit the system environment variables.”
In the System Properties window, click on “Environment Variables.”
Under “System Variables,” find and select the “Path” variable, then click “Edit.”
Click “New” and paste the copied path of the Python directory.
Click “OK” to save the changes.
